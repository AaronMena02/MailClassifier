```python
import pandas as pd
from googletrans import Translator
import nlpaug.augmenter.word as naw
import random

# Cargar dataset original
df = pd.read_csv("dataset_limpio.csv")

# Filtrar solo las columnas que queremos aumentar (por ejemplo 'text')
texts = df["text"].tolist()

translator = Translator()

# Función para back-translation español->italiano->español
def back_translate(text, src='es', mid='it'):
    try:
        # Traduce a idioma intermedio
        translated = translator.translate(text, src=src, dest=mid).text
        # Traduce de regreso a español
        back_translated = translator.translate(translated, src=mid, dest=src).text
        return back_translated
    except Exception as e:
        print(f"Error en traducción: {e}")
        return text

# Inicializar augmenter para sinónimos en español
syn_aug = naw.SynonymAug(aug_src='wordnet', lang='spa')

# Número de aumentos por texto
num_aug_per_text = 2

augmented_texts = []

for text in texts:
    # Back-translation
    bt_text = back_translate(text)
    if bt_text != text:
        augmented_texts.append(bt_text)
    
    # Sinónimos (puede cambiar algunas palabras por sinónimos)
    syn_text = syn_aug.augment(text)
    if syn_text != text:
        augmented_texts.append(syn_text)
    
    # Repetir para más variaciones
    for _ in range(num_aug_per_text - 2):
        # Podemos añadir variación combinada o más sinónimos
        augmented_texts.append(syn_aug.augment(text))

# Crear nuevo dataframe con textos aumentados (sin etiquetas, ojo)
df_aug = pd.DataFrame({"text": augmented_texts})

# Puedes decidir agregar las columnas originales para mantener etiquetas,
# asumiendo que el sentido no cambia

# Por ejemplo, si quieres mantener etiquetas idénticas:
df_aug["sentiment"] = df["sentiment"].sample(len(df_aug), replace=True).reset_index(drop=True)
df_aug["priority"] = df["priority"].sample(len(df_aug), replace=True).reset_index(drop=True)
df_aug["category"] = df["category"].sample(len(df_aug), replace=True).reset_index(drop=True)

# Finalmente, concatenar con el dataset original para balancear
df_final = pd.concat([df, df_aug], ignore_index=True)

# Guardar nuevo dataset aumentado
df_final.to_csv("dataset_limpio_aumentado.csv", index=False)

print(f"Dataset original: {len(df)} filas")
print(f"Datos aumentados: {len(df_aug)} filas")
print(f"Dataset total final: {len(df_final)} filas")
```
